## Diagnostic Analysis

### Clustering (numeric columns only)

- the clustering algorithm will need to **compute distances** in order to group similar data together to **scale** the features we are going to cluster on.

```python
# Drop the non-numeric columns
notnumeric = ['Country']

# cluster columns
df_numeric_only = df_not_null.drop([x for x in notnumeric if x in df_not_null.columns], axis=1)
df_numeric_only.shape
```

```python
 df_numeric_only.columns
```

### Scaling the features/columns

```python
cluster_columns = ['Co2', 'Co2_per_capita', 'Gdp', 'Gdp_per_capita', 'Coal_co2',
       'Cement_co2', 'Flaring_co2', 'Gas_co2', 'Oil_co2',
       'Coal_co2_per_capita', 'Cement_co2_per_capita',
       'Flaring_co2_per_capita', 'Gas_co2_per_capita', 'Oil_co2_per_capita',
       'Cumulative_co2', 'Cumulative_coal_co2', 'Cumulative_cement_co2',
       'Cumulative_flaring_co2', 'Cumulative_gas_co2', 'Cumulative_oil_co2',
       'Annual_co2_fossil_cement', 'Co2_per_gdp']
```

```python
from sklearn import preprocessing 
scaler = preprocessing.MaxAbsScaler() # which will divide each value by the max absolute value for that feature. 
CC_clusters = scaler.fit_transform(df_numeric_only[cluster_columns])
CC_clusters = pd.DataFrame(CC_clusters, columns=cluster_columns)
CC_clusters.shape
```

```python
from sklearn.cluster import KMeans
# will k-means cluster a given data frame into a specified number of clusters.
def kmeans_cluster(df, n_clusters=2):
    model = KMeans(n_clusters=n_clusters, random_state=1)
    clusters = model.fit_predict(df)
    cluster_results = df.copy()
    cluster_results['Cluster'] = clusters
    
    # return a copy of the original data frame with those clusters appended in a column named Cluster.
    return cluster_results
```

```python
# to count the number of samples that fall into each cluster and calculate the cluster means for each feature.
def summarize_clustering(results):
    cluster_size = results.groupby(['Cluster']).size().reset_index()
    cluster_size.columns = ['Cluster', 'Count']
    cluster_means = results.groupby(['Cluster'], as_index=False).mean()
    cluster_summary = pd.merge(cluster_size, cluster_means, on='Cluster')
    return cluster_summary
```

```python
cluster_results = kmeans_cluster(CC_clusters, 3)
cluster_summary = summarize_clustering(cluster_results)
cluster_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cluster_summary[cluster_columns].transpose(), annot=True)
```

```python
cluster_results['Cluster_name'] = ''

# according to heatmap, set the clusters
cluster_results['Cluster_name'][cluster_results['Cluster']==0] = 'Low'
cluster_results['Cluster_name'][cluster_results['Cluster']==2] = 'Medium'
cluster_results['Cluster_name'][cluster_results['Cluster']==1] = 'High'

df_new = df_not_null.reset_index().drop('index', axis=1)
df_new['Cluster_name'] = cluster_results['Cluster_name']
df_new
```

### Visualization

```python
sns.lineplot(x = "Gdp_per_capita", y = "Co2", hue = "Country", style = "Cluster_name", data = df_new)
plt.xticks(rotation = 90)
plt.show()
```

### Interpretation

We clustered our datasets based on any two variables like [var1: high, var2:high] cluster, [var1:high, var2:low] cluster etc .. to see the interesting patterns of the dataset. For example, how many observations which include under high Co2 and high Gdp cluster? which country(s) included in that cluster? etc 
  
We can do the clustering on numeric columns only. So, we removed the country column which is categorial. 

On the rest variables, we need to remove the missing values as well.
After that, we run the KMeans clustering algorithm to see which observation/sample are under which cluster? 

Selected Co2 and Gdp per capita variables to find patterns from datasets.  Found that there are three different patterns in our dataset. 
Cluster 1 has 
- High GDP per capita and high Co2 emission. In cluster 1, only one country included, United State. 
Cluster 2 has
- Medium GDP per capita and medium Co2 emission. United States and Russia.
Cluster 3 has
- Low GDP per capita and Low CO2 emission. Russia, Japan and India

This is how we clustered our dataset based on two main variables, Co2 and Gdp per capita data.
